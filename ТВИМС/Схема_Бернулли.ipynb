# Схема Бернулли. Вариант 6

```python
import math
```

### Задача 1

```python
n = 100000
m = 5800
p = 0.06
q = 1 - p

# Бернулли
logP = (math.log(math.comb(n, m)) +
        m * math.log(p) +
        (n - m) * math.log(q))

P = math.exp(logP)
print(f"P = {P:.6e}")
```

### Задача 2

```python
n = 1200
p = 0.002
m = 0

logP = m * math.log(p) + (n - m) * math.log(1 - p)  
P = math.exp(logP)
print(f"P = {P:.6f}")
```
