# 2.6 Решение алгебраических и трансцендентных уравнений с помощью инструментальных средств

### Задание 1. Найдите методом половинного деления корень транцендетного уравнения lgx-cosx=0 с четырьмя знаками после запятой. Корни отделите графически.

```python
import numpy as np
import matplotlib.pyplot as plt

# Функция
def f(x):
    return np.log10(x) - np.cos(x)

# График
x = np.linspace(0.1, 3, 500)
plt.plot(x, f(x), 'b-')
plt.axhline(0, color='r', ls='--')
plt.grid(True)
plt.title('lg x - cos x = 0')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()

# Метод половинного деления
a, b = 1.3, 1.5  # интервал из графика
for _ in range(30):
    c = (a + b) / 2
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c
    if (b - a) < 1e-5:
        break

print(f"Корень: {c:.4f}")
```

### Задание 2. Отделите графически корни уравнения х^2-3,2х=1. Найдите одним из итерационных методов действительный корень данного уравнения с точностью 10^-4

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2 - 3.2*x - 1

# Находим корень методом половинного деления
a, b = 3.0, 3.5
for _ in range(30):
    c = (a + b) / 2
    if f(a) * f(c) < 0:
        b = c
    else:
        a = c
    if (b - a) < 1e-5:
        break

# График
x = np.linspace(-1, 5, 400)
y = f(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='f(x) = x² - 3.2x - 1')
plt.axhline(0, color='r', linestyle='--', label='y = 0')
plt.axvline(0, color='gray', alpha=0.5)
plt.grid(True, alpha=0.3)
plt.plot(c, 0, 'ro', markersize=8, label=f'Корень: x = {c:.4f}')
plt.plot([a, b], [0, 0], 'g|', markersize=12, label='Интервал')
plt.title('Уравнение x² - 3.2x = 1')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()

print(f"Корень: {c:.4f}")
```

### Задание 3. Найдите наибольший положительный корень уравнения 4х-lnx=4 с точностью 10^-4,используя метод итераций. Корни отделите графически

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    if x <= 0:
        return np.nan
    return 4*x - 3*np.log(x) - 4

def phi(x):
    return (4 + 3*np.log(x)) / 4

# 1. Находим корень методом итераций
x = 2.0
for i in range(30):
    x_new = phi(x)
    if abs(x_new - x) < 1e-4:
        break
    x = x_new

root = x_new

# 2. График
x_vals = np.linspace(0.1, 4, 500)
y_vals = [f(val) for val in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x) = 4x - 3ln x - 4')
plt.axhline(0, color='r', linestyle='--', label='y = 0')
plt.axvline(0, color='gray', alpha=0.5)
plt.grid(True, alpha=0.3)

# Отмечаем корень
plt.plot(root, 0, 'ro', markersize=8, label=f'Корень: x = {root:.4f}')
plt.plot(root, f(root), 'ro', markersize=8)

plt.title('Уравнение 4x - 3ln x = 4 (наибольший положительный корень)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.xlim(0, 4)
plt.ylim(-3, 5)
plt.show()

print(f"Наибольший положительный корень: {root:.4f}")
print(f"Проверка: 4*{root:.4f} - 3ln({root:.4f}) = {4*root - 3*np.log(root):.6f} (должно быть 4)")
```
